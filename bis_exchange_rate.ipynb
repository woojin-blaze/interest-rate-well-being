```python
import json
import requests
import pprint
import re
import pandas as pd

urls = ["https://stats.bis.org/api/v2/data/dataflow/BIS/WS_CBTA/1.0/A.AT+AU+BE+CA+CH+CL+CZ+DE+DK+EE+ES+FI+FR+GB+GR+HU+IE+IL+IT+JP+KR+LT+LU+LV+MX+NL+NO+NZ+PL+PT+SE+SI+SK+TR+US..USD?startPeriod=2014-01-01&endPeriod=2024-07-25&format=sdmx-json"]

data = [json.loads(requests.get(url).content) for url in urls]
#pprint.pprint(json.dumps(data[0])[:1200])
#print(data[0]["data"]["dataSets"][0]["series"]["0:0:0:0:0:0"]["observations"])

country_abbv = data[0]["data"]["structure"]["dimensions"]["series"][1]['values']
country_list = []
for item in country_abbv: 
    country_list.append(item["id"])


new_list = []
for key, value in data[0]["data"]["dataSets"][0]["series"].items():
    #s = re.search('0:(\d+):0:0:0:0',item).group(1)
    #print(s)
    #print(value["observations"])
    alist = []
    for k, val in value["observations"].items():
        alist.append(val[0])
    new_list.append(alist)
    
#print(new_list)
#print(data[0]["data"]["dataSets"][0]["series"]["0:0:0:0:0:0"]["observations"])
total_asset_dict = dict(zip(country_list, new_list))
#pprint.pprint(total_asset_dict)

print(country_abbv)
```

```python
print(total_asset_dict)
```

```python
years = list(range(2014, 2024))
rows = []
for country, values in total_asset_dict.items():
    for year, value in zip(years, values):
        rows.append({'Country': country, 'Year': year, 'Total Asset':round(float(value), 3)})
df = pd.DataFrame(rows)
```
